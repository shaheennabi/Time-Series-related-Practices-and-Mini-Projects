```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
stock_data = yf.download('AAPL', start='2024-01-01', end='2024-12-30')


stock_data.head(10)
```

### Apple Data plot

```python
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Close'], label='Close Price history')
plt.title('Close Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

### ACF, PACF Plot

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plt.figure(figsize=(12,6))

## ACF Plot
plt.subplot(1,2,1)
plot_acf(stock_data['Close'].diff().diff().dropna(), ax=plt.gca(), lags=50)
plt.title('ACF of the Apple Stock')

## PACF Plot
plt.subplot(1,2,2)
plot_pacf(stock_data['Close'].diff().diff().dropna(), ax=plt.gca(), lags=50, method='ywm')
plt.title('PACF of the Apple Stock')

plt.tight_layout()
plt.show

```
