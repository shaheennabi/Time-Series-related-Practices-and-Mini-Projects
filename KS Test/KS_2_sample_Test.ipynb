```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp

## Generating synthetic data for demonstration
np.random.seed(0)
n = 500

## Strictly stationary series: normally distributed random noise
strict_stationary_series = np.random.normal(0,1,n)

### Non-strictly stationary series: changing variance
non_strict_stationary_series = np.concatenate([
    np.random.normal(0,1,n//2),
    np.random.normal(0,2,n//2)
])


## Function to split the series into two halves and perform the K-S test
def ks_test_stationarity(series):
  split = len(series) // 2
  series_first_half = series[:split]
  series_second_half = series[split:]
  stats, p_value = ks_2samp(series_first_half, series_second_half)
  return stats, p_value


## perform the k-s test on strictly stationary series
ks__stat_strict, ks_p_value_strict = ks_test_stationarity(strict_stationary_series)

## perform the k-s test on non-strictly stationary series
ks__stat_non_strict, ks_p_value_non_strict = ks_test_stationarity(non_strict_stationary_series)

## plotting the series vertically with spacing
plt.figure(figsize=(14, 6)) # Adjust figure size for vertical arrangement
plt.subplot(2, 1, 1)
plt.plot(strict_stationary_series)
plt.title('Strictly Stationary Series')
print(f"K-S Statistic for Strictly Stationary Series: {ks__stat_strict}")
print(f"P-Value for Strictly Stationary Series: {ks_p_value_strict}")

plt.subplot(2, 1, 2)
plt.plot(non_strict_stationary_series)
plt.title('Non-Strictly Stationary Series')
print(f"K-S Statistic for Non-Strictly Stationary Series: {ks__stat_non_strict}")
print(f"P-Value for Non-Strictly Stationary Series: {ks_p_value_non_strict}")

plt.tight_layout(pad=3.0) # Add spacing between the plots
plt.show()
```
