```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf
```

```python
stock_data = yf.download('AAPL', start='2024-01-01', end='2024-12-30')


stock_data.head(10)
```

### Apple Data plot

```python
plt.figure(figsize=(10,6))
plt.plot(stock_data.index, stock_data['Close'], label='Close Price history')
plt.title('Close Price History')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

### Smoothing Methods


### Weighted Moving Average

```python
window_size = 28

## calculate Weighted Moving Average (WMA)
weights = np.arange(1, window_size + 1)
stock_data['WMA'] = stock_data['Close'].rolling(window=window_size).apply(lambda x: np.dot(x, weights) / weights.sum(), raw=True)



# plotting
plt.figure(figsize=(10,6))
plt.plot(stock_data['Close'], label='Close Price', color='blue', alpha=0.5)
plt.plot(stock_data['WMA'],  label=f'WMA {window_size}-day', color='orange')

plt.title('Apple Stock Price with WMA')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```
